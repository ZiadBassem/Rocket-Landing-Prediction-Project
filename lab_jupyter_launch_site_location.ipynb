```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.plugins import MarkerCluster

# Initialize map centered around average location
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
spacex_map = folium.Map(location=map_center, zoom_start=5)

# Add a marker cluster to the map
marker_cluster = MarkerCluster().add_to(spacex_map)

# Loop through the dataframe and add markers to the cluster
for i, row in spacex_df.iterrows():
    # Choose color based on launch outcome
    color = 'green' if row['class'] == 1 else 'red'
    
    # Create marker
    marker = folium.CircleMarker(
        location=[row['Lat'], row['Long']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7
    )
    
    # Add popup label
    popup_text = (
        f"<b>Launch Site:</b> {row['Launch Site']}<br>"
        f"<b>Result:</b> {'Success ✅' if row['class']==1 else 'Failure ❌'}<br>"
        f"<b>Lat:</b> {row['Lat']}<br>"
        f"<b>Long:</b> {row['Long']}"
    )
    marker.add_child(folium.Popup(popup_text, max_width=250))
    
    # Add marker to cluster instead of map directly
    marker.add_to(marker_cluster)

# Display the map
spacex_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
import folium
from folium.plugins import MarkerCluster

# Create base map
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
spacex_map = folium.Map(location=map_center, zoom_start=4)

# Create a marker cluster layer
marker_cluster = MarkerCluster().add_to(spacex_map)

# Add each launch to the cluster with popup
for i, row in spacex_df.iterrows():
    # Color coding for success/failure
    color = 'green' if row['class'] == 1 else 'red'
    
    # CircleMarker for each launch
    marker = folium.CircleMarker(
        location=[row['Lat'], row['Long']],
        radius=6,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7
    )
    
    # ✅ Add popup for every launch
    popup_text = (
        f"<b>Launch Site:</b> {row['Launch Site']}<br>"
        f"<b>Result:</b> {'Success ✅' if row['class']==1 else 'Failure ❌'}<br>"
        f"<b>Latitude:</b> {row['Lat']}<br>"
        f"<b>Longitude:</b> {row['Long']}"
    )
    marker.add_child(folium.Popup(popup_text, max_width=250))
    
    # Add marker into the cluster
    marker_cluster.add_child(marker)

# Show map
spacex_map
```

Next, we need to explore and analyze the proximities of launch sites.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster, MousePosition
import math
from IPython.display import display

# ----------------------------
# Haversine Distance
# ----------------------------
def haversine(coord1, coord2):
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    R = 6371 # km
    phi1, phi2 = math.radians(lat1), math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dlambda = math.radians(lon2 - lon1)
    a = math.sin(dphi/2)**2 + math.cos(phi1)*math.cos(phi2)*math.sin(dlambda/2)**2
    return 2 * R * math.atan2(math.sqrt(a), math.sqrt(1 - a))

# ----------------------------
# Load Launch Data
# ----------------------------
spacex_df = spacex_df
launch_sites = spacex_df[['Launch Site','Lat','Long']].drop_duplicates()

# ----------------------------
# Dictionaries for YOUR manual coords
# ----------------------------
shore_coords = {
    "CCAFS LC-40": (28.5623, -80.5678),
    "CCAFS SLC-40": (28.5632, -80.5680),
    "KSC LC-39A": (28.6082, -80.5941),
    "VAFB SLC-4E": (34.6370, -120.6244)
}

rail_coords = {
    "CCAFS LC-40": (28.392, -80.605),
    "CCAFS SLC-40": (28.565, -80.600),
    "KSC LC-39A": (28.630, -80.620),
    "VAFB SLC-4E": (34.6372, -120.6235)
}

highway_coords = {
    "CCAFS LC-40": (28.520, -80.670),
    "CCAFS SLC-40": (28.565, -80.650),
    "KSC LC-39A": (28.628, -80.640),
    "VAFB SLC-4E": (34.6735, -120.5671)
}

city_coords = {
    "CCAFS LC-40": (28.538, -81.379),   # Orlando
    "CCAFS SLC-40": (28.615, -80.807),  # Cocoa/Titusville area
    "KSC LC-39A": (28.628, -81.210),    # Example near Titusville
    "VAFB SLC-4E": (34.639, -120.457)   # Lompoc
}

feature_colors = {
    "Shoreline": "blue",
    "Railway":   "red",
    "Highway":   "orange",
    "City":      "green"
}

# ----------------------------
# Base Map
# ----------------------------
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
spacex_map = folium.Map(location=map_center, zoom_start=4)

# Add cluster and tools
marker_cluster = MarkerCluster().add_to(spacex_map)

# Hover coordinates at bottom-right
MousePosition(
    position="bottomright",
    separator=" | ",
    empty_string="Hover to see coordinates",
    lng_first=True,
    num_digits=6,
    prefix="Coordinates:"
).add_to(spacex_map)

# Click anywhere to show coordinates in popup
folium.LatLngPopup().add_to(spacex_map)

# ----------------------------
# Build Sites
# ----------------------------
distance_records = []

for _, site in launch_sites.iterrows():
    site_name = site["Launch Site"]
    site_coords = (site["Lat"], site["Long"])

    # ========== Clustered Launch Markers ==========
    site_launches = spacex_df[spacex_df['Launch Site'] == site_name]
    for _, row in site_launches.iterrows():
        color = "green" if row["class"] == 1 else "red"
        popup_text = (
            f"<b>Launch Site:</b> {row['Launch Site']}<br>"
            f"<b>Result:</b> {'Success ✅' if row['class']==1 else 'Failure ❌'}<br>"
            f"<b>Date:</b> {row['Date']}<br>"
            f"<b>Payload:</b> {row['Payload']}<br>"
            f"<b>Orbit:</b> {row['Orbit']}<br>"
            f"<b>Lat:</b> {row['Lat']}, <b>Lon:</b> {row['Long']}"
        )
        folium.CircleMarker(
            location=[row['Lat'], row['Long']],
            radius=6,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=popup_text
        ).add_to(marker_cluster)

    # Distance record per site
    record = {"Launch Site": site_name}

    if site_name in shore_coords:
        tgt = shore_coords[site_name]
        dist = haversine(site_coords, tgt)
        record["Shoreline (km)"] = round(dist,2)
        folium.PolyLine([site_coords, tgt], color=feature_colors["Shoreline"], weight=3,
                        popup=f"<b>Shoreline</b>: {dist:.2f} km").add_to(spacex_map)

    if site_name in rail_coords:
        tgt = rail_coords[site_name]
        dist = haversine(site_coords, tgt)
        record["Railway (km)"] = round(dist,2)
        folium.PolyLine([site_coords, tgt], color=feature_colors["Railway"], weight=3,
                        popup=f"<b>Railway</b>: {dist:.2f} km").add_to(spacex_map)

    if site_name in highway_coords:
        tgt = highway_coords[site_name]
        dist = haversine(site_coords, tgt)
        record["Highway (km)"] = round(dist,2)
        folium.PolyLine([site_coords, tgt], color=feature_colors["Highway"], weight=3,
                        popup=f"<b>Highway</b>: {dist:.2f} km").add_to(spacex_map)

    if site_name in city_coords:
        tgt = city_coords[site_name]
        dist = haversine(site_coords, tgt)
        record["City (km)"] = round(dist,2)
        folium.PolyLine([site_coords, tgt], color=feature_colors["City"], weight=3,
                        popup=f"<b>City</b>: {dist:.2f} km").add_to(spacex_map)

    # Equator (table only)
    eq_coords = (0.0, site_coords[1])
    eq_dist = haversine(site_coords, eq_coords)
    record["Equator (km)"] = round(eq_dist,2)

    distance_records.append(record)

# ----------------------------
# Distance DataFrame
# ----------------------------
distance_df = pd.DataFrame(distance_records)

# Show map + dataframe in notebook
display(spacex_map)
display(distance_df)
```
